## Word2Vec

```python
from gensim.models import word2vec
from sklearn.model_selection import train_test_split

label = df['Views']
# Split data to test and train dataset
train, test, train_label, test_label = train_test_split(filtered_data, label, test_size=0.1)

print(len(label))
print(len(train))
print(len(train_label))
print(len(test))
print(len(test_label))


# Create word to vector
sentences = []
for single_title in train['Title'].values:
    sen = []
    split_title = single_title.split(" ")
    for word in split_title:
        if(word != " "):
            sen.append(word)
    sentences.append(sen)

print("Done")
EMBEDDING_DIM = 150
model = word2vec.Word2Vec(sentences, size=EMBEDDING_DIM, window=7, workers=6, min_count=10)

print(len(model.wv.vocab))
print("Done")
```

## LSTM

```python
from keras.callbacks import LambdaCallback
from keras.layers.recurrent import LSTM
from keras.layers.embeddings import Embedding
from keras.layers import Dense, Activation
from keras.models import Sequential
from keras.utils.data_utils import get_file


pretrained_weights = model.wv.syn0
vocab_size, emdedding_size = pretrained_weights.shape

lstm = Sequential()
lstm.add(Embedding(input_dim=vocab_size, output_dim=emdedding_size, 
                    weights=[pretrained_weights]))
lstm.add(LSTM(units=emdedding_size))
lstm.add(Dense(units=vocab_size))
lstm.add(Activation('softmax'))
lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy')

print(pretrained_weights.shape)
```

```python
X_train, X_valid, Y_train, Y_valid = train_test_split(pretrained_weights, train_label, test_size = 0.20, random_state = 36)

embed_dim = 128
lstm_out = 350
batch_size= 1000
#Here we train the Network.

lstm.fit(X_train, Y_train, batch_size = batch_size, epochs = 10,  verbose = 1)

# Measuring score and accuracy on validation set

score,acc = lstm.evaluate(X_valid, Y_valid, verbose = 2, batch_size = batch_size)
print("Logloss score: %.2f" % (score))
print("Validation set Accuracy: %.2f" % (acc))
```
