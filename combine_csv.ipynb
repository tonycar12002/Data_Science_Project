```python
import csv
import pandas as pd


df1 = pd.read_csv("all_videos_1.csv")
df2 = pd.read_csv("all_videos_2.csv")
df3 = pd.read_csv("all_videos_3.csv")
df4 = pd.read_csv("all_videos_4.csv")
df5 = pd.read_csv("all_videos_5.csv")
df6 = pd.read_csv("all_videos_6.csv")
df7 = pd.read_csv("all_videos_7.csv")
```
