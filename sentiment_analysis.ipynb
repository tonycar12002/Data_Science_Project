# Filter out Chinese in topic

```python
# Only get the attributes we need
df1 = dataset.iloc[:, 0]
df2 = dataset.iloc[:, 2:4]

tile_list = dataset.iloc[:, 11]
#print(tile_list)
cop = re.compile("[\u4e00-\u9fa5]")
for i in range (len(tile_list)):
    tile_list[i] = re.sub(cop ,"",tile_list[i])
    if i%1000 == 0:
        print ("Finish " + str(i))


dataset.iloc[:, 11] = tile_list
    



df3 = dataset.iloc[:, 6:7] 
df4 = dataset.iloc[:, 9]
df5 = dataset.iloc[:, 11]
df6 = dataset.iloc[:, 14:-1]

y = dataset.iloc[:, 12]
data = pd.concat([df1, df2, df3, df4, df5, df6, y], axis=1, join_axes=[df1.index])

print(data.shape)
data
```

# Sentiment analysis and save to csv

```python
import importlib
import sys
importlib.reload(sys)
import sys
from reactionrnn import reactionrnn
sys.getdefaultencoding()
react = reactionrnn()

```

```python
sentiment = pd.DataFrame(np.zeros((dataset.shape[0],5),dtype=float), columns=['haha', 'love', 'angry', 'wow', 'sad'])
dataset.iloc[:, 12]
topic_list = dataset.iloc[:, 12].values
#print (topic_list)
for i in range(len(topic_list)):
    try:
        pred_result = react.predict(topic_list[i])
        sentiment.iloc[i,0] = pred_result["haha"]
        sentiment.iloc[i,1] = pred_result["love"]
        sentiment.iloc[i,2] = pred_result["angry"]
        sentiment.iloc[i,3] = pred_result["wow"]
        sentiment.iloc[i,4] = pred_result["sad"]
    except:
        sentiment.iloc[i,0] = 1
        sentiment.iloc[i,1] = 0
        sentiment.iloc[i,2] = 0
        sentiment.iloc[i,3] = 0
        sentiment.iloc[i,4] = 0
        print ("exception")
    if i%1000 == 0:
        print ("Finish " + str(i))
```

```python
new_data = pd.concat([dataset, sentiment], axis=1, join_axes=[dataset.index])
new_data
```
